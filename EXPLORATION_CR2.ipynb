## 3. 데이터 전처리

reviews_per_month 는 월에 리뷰가 몇개정도 달렸는지 알려주는 지표다. 리뷰가 아얘 없는 경우 nan값이 입력된 것 같다. 0으로 바꿔도 무방하겠다.

```python
#리뷰 데이터가 없는 것을 리뷰 0으로 바꿈.
train['reviews_per_month'] = train['reviews_per_month'].fillna(0)
test['reviews_per_month'] = test['reviews_per_month'].fillna(0)
```

last_review 는 채우기 힘들다고 판단.. 결측치도 20% 정도로 너무 많아서 삭제하기로 결정.

```python
# 결측치 컬럼 삭제 (last_review)
train = train.drop('last_review', axis=1)
test = test.drop('last_review', axis=1)
```

이름과 호스트이름은 어찌해야하나.. 일단 보류...

```python
train.corr(method='pearson')
```

룸데이터는 3가지로 나뉜다. 개수가 적으니 원핫인코딩을 하면 좋겠다. 

```python
# 원핫인코딩 

train_onehot = train[['room_type']].copy()
train_onehot = pd.get_dummies(train_onehot)
train_onehot

test_onehot = test[['room_type']].copy()
test_onehot = pd.get_dummies(test_onehot)
test_onehot
```

```python
#원핫인코딩 데이터 붙이기
train = pd.concat([train,train_onehot], axis=1)
train = train.drop(labels='room_type', axis=1)
train


test = pd.concat([test,test_onehot], axis=1)
test = test.drop(labels='room_type', axis=1)
test
```

데이터가 너무 세분되어있어서 레이블 인코딩을 시도해보겠다.

우선 결측치를 제거한다.

```python
train[['name']] = train[['name']].fillna('noroomname')
test[['name']] = test[['name']].fillna('noroomname')

train[['host_name']] = train[['host_name']].fillna('noname')
test[['host_name']] = test[['host_name']].fillna('noname')
```

```python
train.isnull().sum()
```

미연의 사태를 대비해 결측치를 한번 더 날려보자.

```python
train = train.fillna(0)
test = test.fillna(0)
```

이름과 호스트 레이블 인코딩 시도

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
train['name'] = le.fit_transform(train['name'])
test['name'] = le.fit_transform(test['name'])


train['host_name'] = le.fit_transform(train['host_name'])
test['host_name'] = le.fit_transform(test['host_name'])
```

종류가 많으니 레이블 인코딩 시도

```python
#네이버후드 레이블인코더
le = LabelEncoder()
train['neighbourhood'] = le.fit_transform(train['neighbourhood'])
test['neighbourhood'] = le.fit_transform(test['neighbourhood'])
```

```python
train['neighbourhood_group'].value_counts()
```

네이버후드 그룹은 컬럼이 적으니 원핫인코딩 시도

```python
#네이버후드 그룹 원핫 인코딩
train_onehot_neighbor_group = train[['neighbourhood_group']].copy()
train_onehot_neighbor_group = pd.get_dummies(train_onehot_neighbor_group)
train_onehot_neighbor_group

# train_up['neighbourhood_group'] = 

#네이버후드 그룹, 원핫인코딩한거 upper에 붙이기!
train = pd.concat([train,train_onehot_neighbor_group], axis=1)
train = train.drop(labels='neighbourhood_group', axis=1)
train




#네이버후드 그룹 원핫 인코딩
test_onehot_neighbor_group = test[['neighbourhood_group']].copy()
test_onehot_neighbor_group = pd.get_dummies(test_onehot_neighbor_group)
test_onehot_neighbor_group

# test_up['neighbourhood_group'] = 

#네이버후드 그룹, 원핫인코딩한거 upper에 붙이기!
test = pd.concat([test,test_onehot_neighbor_group], axis=1)
test = test.drop(labels='neighbourhood_group', axis=1)
test
```
